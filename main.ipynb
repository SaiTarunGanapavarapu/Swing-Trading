## A Note on Data Anomalies

Two rows in this output illustrate real-world data caveats rather than code bugs. HDFCBANK.NS shows NaN for EV/EBITDA and Net Debt/EBITDA because these multiples are not structurally compatible with financial institutions: banks do not operate on traditional EBITDA and their balance sheets are deposit-driven, so standard enterprise valuation ratios can break. INFY.NS shows an EV/EBITDA outlier (1054.633), which is likely a yfinance data-quality issue (for example stale or inconsistent denominator inputs) rather than an economically meaningful valuation multiple. This is exactly why the screener uses a multi-factor composite framework with scoring guardrails instead of relying on any single raw metric; despite the outlier, INFY.NS still carries a reasonable valuationScore (~2.9), indicating the composite system is absorbing noisy data as intended.
